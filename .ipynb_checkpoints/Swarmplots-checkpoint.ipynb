```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.swarmplot(cars.mpg);
```

```python
sns.swarmplot(x = cars.cylinders, y = cars.mpg);
```

```python
sns.swarmplot(x = cars.cylinders, y = cars.mpg, hue = cars.origin);
```

```python
sns.swarmplot(x = cars.cylinders, y = cars.mpg, hue = cars.origin,
             dodge = True, linewidth = .75, size = 4);
```

```python
#Overlay on boxplots or violin plots
```

```python
import numpy as np
```

```python
sns.boxplot(x = cars.cylinders, y = cars.mpg, whis=np.inf);
sns.swarmplot(x = cars.cylinders, y = cars.mpg, size = 4.5, color = 'black', alpha = 0.5);
```
